```python
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import img_to_array, load_img
from skimage.color import lab2rgb
import cv2
import os
import tensorflow as tf
import matplotlib.pyplot as plt

# Custom layer definition for loading the model
class ReshapeAndResizeLayer(tf.keras.layers.Layer):
    def call(self, inputs):
        resized = tf.image.resize(inputs, [100, 100])
        return resized

# Load the saved model
model = load_model('colorization_model.keras', custom_objects={"ReshapeAndResizeLayer": ReshapeAndResizeLayer})

# Define paths for input and output
input_folder = './Data/Test/'
output_folder = './Data/Colorized/'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Process each image in the input folder
for img_name in os.listdir(input_folder):
    img_path = os.path.join(input_folder, img_name)

    # Load the grayscale image
    img = load_img(img_path, color_mode="grayscale")
    img_array = img_to_array(img) / 255  # Normalize to [0, 1]
    original_shape = img_array.shape[:2]  # Save original resolution

    # Prepare the input for the model
    resized_input = cv2.resize(img_array, (100, 100))
    X = np.expand_dims(resized_input, axis=2)  # Add channel dimension
    X = np.reshape(X, (1, 100, 100, 1))   # Batch dimension

    # Predict the colorized output
    output = model.predict(X)
    output = np.reshape(output, (100, 100, 2))  # Reshape to match LAB format
    output = cv2.resize(output, (original_shape[1], original_shape[0]))  # Resize to original resolution

    # Prepare the LAB image
    outputLAB = np.zeros((original_shape[0], original_shape[1], 3))
    outputLAB[:, :, 0] = img_array.squeeze() * 100  # L channel
    outputLAB[:, :, 1:] = (output * [255, 255]) - [128, 128]  # AB channels

    # Convert LAB to RGB
    rgb_image = lab2rgb(outputLAB)

    # Save the colorized image
    output_image_path = os.path.join(output_folder, os.path.splitext(img_name)[0] + '_colorized.png')
    plt.imsave(output_image_path, rgb_image)

    # Display grayscale and colorized images side by side
    plt.figure(figsize=(12, 6))
    # Display original grayscale image
    plt.subplot(1, 2, 1)
    plt.title('Black & White')
    plt.imshow(img_array.squeeze(), cmap='gray', extent=(0, original_shape[1], original_shape[0], 0))
    plt.axis('off')
    # Display colorized image
    plt.subplot(1, 2, 2)
    plt.title('Colorized')
    plt.imshow(rgb_image, extent=(0, original_shape[1], original_shape[0], 0))
    plt.axis('off')
    plt.show()

    print(f"Colorized image saved at: {output_image_path}")

print("All images processed and colorized.")
```
